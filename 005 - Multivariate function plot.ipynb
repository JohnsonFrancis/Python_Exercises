# This script will plot the mathematical multivariate functions to a graph

```python
from math import *
from numpy import *
import matplotlib.pyplot as plt
```

```python
def plotxy():
    print('Plot works with standard python operators, math & numpy functons.')
    print('Eg. z=x**3+2*y**2+3*x+9   ,   z=pi*x**2*y  ,  z=log10(x+y)   ,   z=exp(x/y)')
    f=input('PLease enter the function. z=')
    x=arange(-10,10,0.00001)
    y=arange(-100,100,0.0001)
    z=eval(f)
    # syntax for 3-D projection 
    ax=plt.axes(projection ='3d') 
    ax.plot3D(x,y,z)
    ax.set_title('z='+f)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.show()
```

```python
plotxy()
```
