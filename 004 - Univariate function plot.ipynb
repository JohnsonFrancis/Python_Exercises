# This script will plot the mathematical univariate functions to a graph

```python
from math import *
from numpy import *
import matplotlib.pyplot as plt
```

```python
print('Plot works with standard python operators, math & numpy functons.')
print('Eg. y=x**3+2x**2+3*x+9   ,   y=pi*x**2  ,  y=log10(x)   ,   y=exp(x)')
f=input('PLease enter the function. y=')
x=arange(-10,10,0.00001)
y=eval(f)
plt.plot(x,y)
plt.title('y='+f)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
